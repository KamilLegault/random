```python
from torchvision.utils import make_grid
```

```python
data_path = "Downloads/train/"
num_workers = 2
batchSize = 128
img_size = 64
latent_dim = 100
n_gen = 32
n_disc = 32
n_epochs = 25
lr = 0.0002
beta = 0.5
use_cuda = True
num_gpu = 1
path_gen = ""
path_disc = ""
path_output = "output"
nc = 3
try:
    os.makedirs(path_output)
except OSError:
    pass
```

```python
dataset = dset.ImageFolder(root=data_path,
                               transform=transforms.Compose([
                                   transforms.Resize(img_size),
                                   transforms.CenterCrop(img_size),
                                   transforms.ToTensor(),
                                   transforms.Normalize((0.5, 0.5, 0.5), (0.5, 0.5, 0.5)),]))
```

```python
dataloader = torch.utils.data.DataLoader(dataset, batch_size=batchSize,shuffle=True, num_workers=int(num_workers))
```

```python
# custom weights initialization called on netG and netD
def weights_init(m):
    classname = m.__class__.__name__
    if classname.find('Conv') != -1:
        m.weight.data.normal_(0.0, 0.02)
    elif classname.find('BatchNorm') != -1:
        m.weight.data.normal_(1.0, 0.02)
        m.bias.data.fill_(0)
```

```python
class _netG(nn.Module):
    def __init__(self):
        super(_netG, self).__init__()
        self.main = nn.Sequential(
            # input is Z, going into a convolution
            nn.ConvTranspose2d(     latent_dim, n_gen * 8, 4, 1, 0, bias=False),
            nn.BatchNorm2d(n_gen * 8),
            nn.ReLU(True),
            # state size. (ngf*8) x 4 x 4
            nn.ConvTranspose2d(n_gen * 8, n_gen * 4, 4, 2, 1, bias=False),
            nn.BatchNorm2d(n_gen * 4),
            nn.ReLU(True),
            # state size. (ngf*4) x 8 x 8
            nn.ConvTranspose2d(n_gen * 4, n_gen * 2, 4, 2, 1, bias=False),
            nn.BatchNorm2d(n_gen * 2),
            nn.ReLU(True),
            # state size. (ngf*2) x 16 x 16
            nn.ConvTranspose2d(n_gen * 2,     n_gen, 4, 2, 1, bias=False),
            nn.BatchNorm2d(n_gen),
            nn.ReLU(True),
            # state size. (ngf) x 32 x 32
            nn.ConvTranspose2d(    n_gen,      nc, 4, 2, 1, bias=False),
            nn.Tanh()
            # state size. (nc) x 64 x 64
        )

    def forward(self, input):
            output = self.main(input)
            return output
```

```python
netG_v = _netG()
netG_v.apply(weights_init)
```

```python

class _netD(nn.Module):
    def __init__(self):
        super(_netD, self).__init__()
        self.main = nn.Sequential(
            # input is (nc) x 64 x 64
            nn.Conv2d(nc, n_disc, 4, 2, 1, bias=False),
            nn.LeakyReLU(0.2, inplace=True),
            # state size. (ndf) x 32 x 32
            nn.Conv2d(n_disc, n_disc * 2, 4, 2, 1, bias=False),
            nn.BatchNorm2d(n_disc * 2),
            nn.LeakyReLU(0.2, inplace=True),
            # state size. (ndf*2) x 16 x 16
            nn.Conv2d(n_disc * 2, n_disc * 4, 4, 2, 1, bias=False),
            nn.BatchNorm2d(n_disc * 4),
            nn.LeakyReLU(0.2, inplace=True),
            # state size. (ndf*4) x 8 x 8
            nn.Conv2d(n_disc * 4, n_disc * 8, 4, 2, 1, bias=False),
            nn.BatchNorm2d(n_disc * 8),
            nn.LeakyReLU(0.2, inplace=True),
            # state size. (ndf*8) x 4 x 4
            nn.Conv2d(n_disc * 8, 1, 4, 1, 0, bias=False),
            nn.Sigmoid()
        )

    def forward(self, input):
        output = self.main(input)
        return output.view(-1, 1).squeeze(1)
```

```python
netD_v = _netD()
netD_v.apply(weights_init)
print(netD_v)
```

```python
criterion = nn.BCELoss()

input = torch.FloatTensor(batchSize, 3, img_size, img_size)
noise = torch.FloatTensor(batchSize, latent_dim, 1, 1)
fixed_noise = torch.FloatTensor(batchSize, latent_dim, 1, 1).normal_(0, 1)
label = torch.FloatTensor(batchSize)
real_label = 1
fake_label = 0

if use_cuda:
    netD_v.cuda()
    netG_v.cuda()
    criterion.cuda()
    input, label = input.cuda(), label.cuda()
    noise, fixed_noise = noise.cuda(), fixed_noise.cuda()

fixed_noise = Variable(fixed_noise)

# setup optimizer
optimizerD = optim.Adam(netD_v.parameters(), lr=lr, betas=(beta, 0.999))
optimizerG = optim.Adam(netG_v.parameters(), lr=lr, betas=(beta, 0.999))
```

### 2.

```python
for epoch in range(10):
    for i, data in enumerate(dataloader, 0):
        ############################
        # (1) Update D network: maximize log(D(x)) + log(1 - D(G(z)))
        ###########################
        # train with real
        netD_v.zero_grad()
        real_cpu, _ = data
        batch_size = real_cpu.size(0)
        if use_cuda:
            real_cpu = real_cpu.cuda()
        input.resize_as_(real_cpu).copy_(real_cpu)
        label.resize_(batch_size).fill_(real_label)
        inputv = Variable(input)
        labelv = Variable(label)

        output = netD_v(inputv)
        errD_real = criterion(output, labelv)
        errD_real.backward()
        D_x = output.data.mean()

        # train with fake
        noise.resize_(batch_size, latent_dim, 1, 1).normal_(0, 1)
        noisev = Variable(noise)
        fake = netG_v(noisev)
        labelv = Variable(label.fill_(fake_label))
        output = netD_v(fake.detach())
        errD_fake = criterion(output, labelv)
        errD_fake.backward()
        D_G_z1 = output.data.mean()
        errD = errD_real + errD_fake
        optimizerD.step()

        ############################
        # (2) Update G network: maximize log(D(G(z)))
        ###########################
        if(i%1 == 0):
            netG_v.zero_grad()
            labelv = Variable(label.fill_(real_label))  # fake labels are real for generator cost
            output = netD_v(fake)
            errG = criterion(output, labelv)
            errG.backward()
            D_G_z2 = output.data.mean()
            optimizerG.step()

        if i % 100 == 0:
            print('[%d/%d][%d/%d] Loss_D: %.4f Loss_G: %.4f D(x): %.4f D(G(z)): %.4f / %.4f'
              % (epoch, 5, i, len(dataloader),
                 errD.data[0], errG.data[0], D_x, D_G_z1, D_G_z2))

    #show result
    vutils.save_image(fake.data,
            '%s/fake_samples_epoch_%03d.png' % ("output", epoch),
            normalize=True)
    # do checkpointing
    torch.save(netG_v.state_dict(), '%s/netG_epoch_%d.pth' % (path_output, epoch))
    torch.save(netD_v.state_dict(), '%s/netD_epoch_%d.pth' % (path_output, epoch))
```

```python
netD_w = _netD()
netD_w.apply(weights_init)
netG_w = _netG()
netG_w.apply(weights_init)

optimizerD_w = optim.RMSprop(netD_w.parameters(), lr=2e-4)
optimizerG_w = optim.RMSprop(netG_w.parameters(), lr=2e-4)

if use_cuda:
    netD_w.cuda()
    netG_w.cuda()
```

```python
for epoch in range(20):
    j = 0
    for i, data in enumerate(dataloader, 0):
        ############################
        # (1) Update D network: maximize log(D(x)) + log(1 - D(G(z)))
        ###########################
        # train with real
        netD_w.zero_grad()
        real_cpu, _ = data
        batch_size = real_cpu.size(0)
        if use_cuda:
            real_cpu = real_cpu.cuda()
        input.resize_as_(real_cpu).copy_(real_cpu)
        label.resize_(batch_size).fill_(real_label)
        inputv = Variable(input)
        labelv = Variable(label)

        output = netD_w(inputv)

        #errD_real = torch.mean((output - 1)**2)
        errD_real = torch.mean(output) #criterion(output, labelv)
        #errD_real.backward()
        D_x = output.data.mean()

        # train with fake
        noise.resize_(batch_size, latent_dim, 1, 1).normal_(0, 1)
        noisev = Variable(noise)
        fake = netG_w(noisev)
        labelv = Variable(label.fill_(fake_label))
        output = netD_w(fake.detach())
        #errD_fake = torch.mean(output**2)
        errD_fake = torch.mean(output)#torch.mean(output**2)#criterion(output, labelv)
        #errD_fake.backward()
        D_G_z1 = output.data.mean()
        errD = (errD_real - errD_fake)#0.5*(errD_real + errD_fake)
        #errD = 0.5 * (errD_real+errD_fake)
        errD.backward()
        optimizerD_w.step()
        
        for p in netD_w.parameters():
            p.data.clamp_(-0.05, 0.05)

        ############################
        # (2) Update G network: maximize log(D(G(z)))
        ###########################
        if(i%2 ==0):
            netG_w.zero_grad()
            labelv = Variable(label.fill_(real_label))  # fake labels are real for generator cost
            output = netD_w(fake)
            #errG = 0.5 * torch.mean((output - 1)**2) #criterion(output, labelv)
            errG = torch.mean(output)
            errG.backward()
            D_G_z2 = output.data.mean()
            optimizerG_w.step()
        j = j+1
    
        if i % 100 == 0:
            print('[%d/%d][%d/%d] Loss_D: %.4f Loss_G: %.4f D(x): %.4f D(G(z)): %.4f / %.4f'
              % (epoch, 15, i, len(dataloader),
                 errD.data[0], errG.data[0], D_x, errD_fake, errD_real))
    torch.save(netG_v.state_dict(), '%s/netG_w_epoch_%d.pth' % (path_output, epoch))
    torch.save(netD_v.state_dict(), '%s/netD_w_epoch_%d.pth' % (path_output, epoch))
```

```python
def train_model(modelG, modelD, latent_dim, epochs=2):
    print("Latent DIM:{}".format(latent_dim))
    criterion = nn.BCELoss().cuda()
    input = torch.FloatTensor(batchSize, 3, img_size, img_size).cuda()
    noise = torch.FloatTensor(batchSize, latent_dim, 1, 1).cuda()
    fixed_noise = torch.FloatTensor(batchSize, latent_dim, 1, 1).normal_(0, 1).cuda()
    fixed_noise = Variable(fixed_noise)
    label = torch.FloatTensor(batchSize).cuda()
    real_label = 1
    fake_label = 0
    optimizerD = optim.Adam(modelD.parameters(), lr=lr, betas=(beta, 0.999))
    optimizerG = optim.Adam(modelG.parameters(), lr=lr, betas=(beta, 0.999))

    for epoch in range(epochs):
        for i, data in enumerate(dataloader, 0):
            ############################
            # (1) Update D network: maximize log(D(x)) + log(1 - D(G(z)))
            ###########################
            # train with real
            modelD.zero_grad()
            real_cpu, _ = data
            batch_size = real_cpu.size(0)
            if use_cuda:
                real_cpu = real_cpu.cuda()
            input.resize_as_(real_cpu).copy_(real_cpu)
            label.resize_(batch_size).fill_(real_label)
            inputv = Variable(input)
            labelv = Variable(label)

            output = modelD(inputv)
            errD_real = criterion(output, labelv)
            errD_real.backward()
            D_x = output.data.mean()

            # train with fake
            noise.resize_(batch_size, latent_dim, 1, 1).normal_(0, 1)
            noisev = Variable(noise)
            fake = modelG(noisev)
            labelv = Variable(label.fill_(fake_label))
            output = modelD(fake.detach())
            errD_fake = criterion(output, labelv)
            errD_fake.backward()
            D_G_z1 = output.data.mean()
            errD = errD_real + errD_fake
            optimizerD.step()

            ############################
            # (2) Update G network: maximize log(D(G(z)))
            ###########################
            if(i%1 == 0):
                modelG.zero_grad()
                labelv = Variable(label.fill_(real_label))  # fake labels are real for generator cost
                output = modelD(fake)
                errG = criterion(output, labelv)
                errG.backward()
                D_G_z2 = output.data.mean()
                optimizerG.step()

            if i % 100 == 0:
                print('[%d/%d][%d/%d] Loss_D: %.4f Loss_G: %.4f D(x): %.4f D(G(z)): %.4f / %.4f'
                  % (epoch, 5, i, len(dataloader),
                     errD.data[0], errG.data[0], D_x, D_G_z1, D_G_z2))
                
    return modelG,modelD
```
